```python
import pandas as pd
```

```python
file_name = None  # if set to None, latest file with results will be selected

results = (
    pd.read_parquet("20231213_01_13_17_retrain_freq_60.parquet")
    .assign(horizon=lambda df: (df.index - df["forecast_date"]).dt.days)
)
results.groupby(["model", "horizon"]).size()
```

```python
(
    results
    .assign(
        ae=lambda df: df["y_pred"].subtract(df["y_test"]).abs(),
        horizon=lambda df: (df.index - df["forecast_date"]).dt.days
    )
    .groupby(["model", "horizon"])
    ["ae"]
    .mean()
)
```

```python
(
    results
    .assign(
        ae=lambda df: df["y_pred"].subtract(df["y_test"]).abs(),
        horizon=lambda df: (df.index - df["forecast_date"]).dt.days
    )
    .groupby("model")
    ["ae"]
    .mean()
)
```
